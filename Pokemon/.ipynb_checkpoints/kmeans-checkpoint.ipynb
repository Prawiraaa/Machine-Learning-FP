```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./pokemonData/Pokemon.csv')
print(f"df length: {len(df)}")
df.head()
```

```python
#features = ["Total", "HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]
datall = df.loc[:, ["Total", "HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]]

from sklearn.preprocessing import StandardScaler
def poKmeans(features, clusters):
    X = df.loc[:, features]
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    kmeans = KMeans(n_clusters=clusters, random_state=37)
    labels = kmeans.fit(X_scaled)
    df['Cluster'] = kmeans.fit_predict(X_scaled)
    sns.pairplot(df, hue='Cluster', vars=features, palette='viridis')
    plt.show()
    print(labels.labels_)

    
    
```

Base stats : 3 Clusters<br>
>BST divides:<br>
>>0 - 400<br>
>>400 - 600<br>
>>600 - beyond<br>


```python
poKmeans(['Attack', 'Speed'], 4)
```

```python
poKmeans(['Defense', 'Sp. Def'], 4)
```
